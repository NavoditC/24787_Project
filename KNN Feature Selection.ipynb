```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import warnings
from sklearn.preprocessing import StandardScaler

warnings.filterwarnings("ignore")

# Reading the dataset
dataset = pd.read_csv("wdbc.data", sep=",").values

#Normalizing all Data by feature
normDataset = np.empty(dataset.shape)
normDataset[:,0:1] = dataset[:,0:1]
print(np.mean(dataset[:,2:2+1]))

print(np.std(dataset[:,2:2+1]))
for i in range(29):
    index = i+2
    normDataset[:,index:index+1] = np.divide(dataset[:,index:index+1]-np.mean(dataset[:,index:index+1]),np.std(dataset[:,index:index+1]))

# Forming a data frame 
attributes = ["ID","Diagnosis","mean_radius","mean_texture","mean_perimeter","mean_area","mean_smoothness","mean_compactness","mean_concavity","mean_concave_points","mean_symmetry","mean_fractal_dimension","se_radius","se_texture","se_perimeter","se_area","se_smoothness","se_compactness","se_concavity","se_concave_points","se_symmetry","se_fractal_dimension",
            "worst_radius","worst_texture","worst_perimeter","worst_area","worst_smoothness","worst_compactness","worst_concavity","worst_concave_points","worst_symmetry","worst_fractal_dimension"]
data = pd.DataFrame(data=dataset,columns=attributes)
data.set_index("ID", inplace=True)

# Forming a data frame with the normalized data
normData = pd.DataFrame(data=normDataset,columns=attributes)
normData.set_index("ID", inplace=True)



# Displaying the diagnosis(M = malignant, B = benign) and the 30 real-valued attributes
display(data.head())
display(normData.head())
```

```python
# Label-1 indicates malignant and Label-0 indicates benign 
data["Diagnosis"] = data["Diagnosis"].replace("M",1)
data["Diagnosis"] = data["Diagnosis"].replace("B",0)
data = data.apply(pd.to_numeric, errors="ignore")

#overwrite with normalized data
replace_label_array = ["mean_radius","mean_texture","mean_perimeter","mean_area","mean_smoothness","mean_compactness","mean_concavity","mean_concave_points","mean_symmetry","mean_fractal_dimension",
                       "se_radius","se_texture","se_perimeter","se_area","se_smoothness","se_compactness","se_concavity","se_concave_points","se_symmetry","se_fractal_dimension",
                        "worst_radius","worst_texture","worst_perimeter","worst_area","worst_smoothness","worst_compactness","worst_concavity","worst_concave_points","worst_symmetry","worst_fractal_dimension"]

data[replace_label_array] = normData[replace_label_array]  #<<<< comment out to use non-normalized data

dataset[:,2:]=normDataset[:,2:]

# X contains only the first 10 features i.e only the mean values of the attributes not their standard error or maximum values

# Mean
X1 = dataset[:,2:12]
X1 = X1.astype("float64")
X1 = np.concatenate([np.ones((X1.shape[0],1)),X1], axis = 1)
x1Labels = attributes[2:12]

# Standard error
X2 = dataset[:,12:22]
X2 = X2.astype("float64")
X2 = np.concatenate([np.ones((X2.shape[0],1)),X2], axis = 1)
x2Labels = attributes[12:22]

# Maximum
X3 = dataset[:,22:32]
X3 = X3.astype("float64")
X3 = np.concatenate([np.ones((X3.shape[0],1)),X3], axis = 1)
x3Labels = attributes[22:32]

# Mean and Standard error
X4 = dataset[:,2:22]
X4 = X4.astype("float64")
X4 = np.concatenate([np.ones((X4.shape[0],1)),X4], axis = 1)
x4Labels = attributes[2:22]

# Standard error and Maximum
X5 = dataset[:,12:]
X5 = X5.astype("float64")
X5 = np.concatenate([np.ones((X5.shape[0],1)),X5], axis = 1)
x5Labels = attributes[12:]

# Mean and Maximum
X6 = np.concatenate([X1,dataset[:,22:32]],axis=1)
X6 = X6.astype("float64")
x6Labels = np.concatenate([x1Labels,x3Labels])
                     
# Mean, Standard error and Maximum
X7 = dataset[:,2:]
X7 = X7.astype("float64")
X7 = np.concatenate([np.ones((X7.shape[0],1)),X7], axis = 1)
x7Labels = attributes[2:]


labels = data["Diagnosis"].values.reshape(-1,1)
data.drop("Diagnosis", axis = 1, inplace = True)
final_data1 = np.concatenate([X1,labels],axis=1)
final_data2 = np.concatenate([X2,labels],axis=1)
final_data3 = np.concatenate([X3,labels],axis=1)
final_data4 = np.concatenate([X4,labels],axis=1)
final_data5 = np.concatenate([X5,labels],axis=1)
final_data6 = np.concatenate([X6,labels],axis=1)
final_data7 = np.concatenate([X7,labels],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report
np.random.seed(0)


def evaluate(final_data, labelSubset):
    m,n = final_data.shape
    #print( final_data.shape )
    #print( final_data )
    
    overall_optimal_nearest_neighbors = -1
    max_f1 = -1
    index1 = -1
    index2 = -2
    accuracy_final = 0
    
    
    
    #index 0 is the diagnosis
    
    for iteration_Fold in range(5): # five fold
        np.random.shuffle(final_data)
        for i in range(1,n-1):
            for j in range(i+1,n-1):

                x_train = np.concatenate((final_data[0:int(0.8*m),i].reshape(-1,1),final_data[0:int(0.8*m),j].reshape(-1,1)),axis=1)
                y_train = final_data[0:int(0.8*m),n-1]
                x_test = np.concatenate((final_data[int(0.8*m):,i].reshape(-1,1),final_data[int(0.8*m):,j].reshape(-1,1)),axis=1)
                y_test  = final_data[int(0.8*m):,n-1]

                neighbors = 5
                knn = KNeighborsClassifier(n_neighbors = neighbors)
                knn.fit(x_train,y_train)
                predictions = knn.predict(x_test)        

                p = classification_report(y_test,predictions,output_dict = True).get('macro avg').get('precision')
                r = classification_report(y_test,predictions,output_dict = True).get('macro avg').get('recall')
                f1 = (2*r*p)/(r+p)
                accuracy_meas = classification_report(y_test,predictions,output_dict = True).get('accuracy')
                
                if(1.0 == f1):
                    print(str(labelSubset[i-1]) + " & " + str(labelSubset[j-1]) + "\t\t F1: " + str(f1))
                
                
#                 if round(local_max_f1, 5)>round(f1, 5):
#                     local_max_f1 = f1
#                     local_optimal_nn=neighbors
                
                if f1>max_f1:
                        max_f1 = f1 #local_max_f1
                        #overall_optimal_nearest_neighbors = local_optimal_nn
                        index1 = i -1
                        index2 = j -1
                        accuracy_final=  accuracy_meas
    print("----- Final stats for 2 Selected Features-----")
    print("F1 = {}\n".format(max_f1))
    print("accuracy = {}\n".format(accuracy_final))
    print("Indices\n{} and {}\n".format(index1,index2))
    print("Feature Combination:\n{}\n{}\n".format(labelSubset[index1],labelSubset[index2]))
    
    
    
```

```python
# Maximum
evaluate(final_data3, x3Labels)
```

```python
np.random.seed(0)
import time

def evaluateFractTime(final_data, labelSubset):
    m,n = final_data.shape
    print( final_data.shape )
    #print( final_data )
    
    
    overall_optimal_nearest_neighbors = -1
    max_f1 = -1
    i = 0+1
    j = 1+1
    k = 3+1
    l = 4+1
    final_acc = 0
    
    elapsedTime = 0
    #index 0 is the diagnosis
    
    
    for iteration_Fold in range(5): # five fold
        np.random.shuffle(final_data)
        
        
        x_train = np.concatenate((final_data[0:int(0.8*m),i].reshape(-1,1),final_data[0:int(0.8*m),j].reshape(-1,1),final_data[0:int(0.8*m),k].reshape(-1,1),final_data[0:int(0.8*m),l].reshape(-1,1)),axis=1)
        y_train = final_data[0:int(0.8*m),n-1]
        x_test = np.concatenate((final_data[int(0.8*m):,i].reshape(-1,1),final_data[int(0.8*m):,j].reshape(-1,1),final_data[int(0.8*m):,k].reshape(-1,1),final_data[int(0.8*m):,l].reshape(-1,1)),axis=1)
        y_test  = final_data[int(0.8*m):,n-1]

        neighbors = 5

        knn = KNeighborsClassifier(n_neighbors = neighbors)

        start = time.time() # <<<<< start time
        knn.fit(x_train,y_train)
        predictions = knn.predict(x_test)        
        stop = time.time() # <<<<< end  time
        elapsedTime = elapsedTime + ((stop - start)*1000) # <<<<< add elapsed time
        

        p = classification_report(y_test,predictions,output_dict = True).get('macro avg').get('precision')
        r = classification_report(y_test,predictions,output_dict = True).get('macro avg').get('recall')
        f1 = (2*r*p)/(r+p)            
        accuracy_meas = classification_report(y_test,predictions,output_dict = True).get('accuracy')
        accuracy_meas = (sum([1 if y_test[i]==predictions[i] else 0 for i in range(y_test.shape[0])])/y_test.shape[0])*100


    #overall_optimal_nearest_neighbors = local_optimal_nn
    
    final_acc = accuracy_meas

    print("----- Final stats for one run-----")
    print("F1 = {}\n".format(f1))
    print("accuracy = {}\n".format(final_acc))
    print("Indices\n{},{},{},and {}\n".format(i-1,j-1,k-1,l-1))
    print("Total Time = {} ms\n".format(elapsedTime/5))
    print("Feature Combination:\n{}\n{}\n{}\n{}\n".format(labelSubset[i-1],labelSubset[j-1],labelSubset[k-1],labelSubset[l-1]))

    
    
```
